# Python Basics: Data visualization

Data visualization is a vast field that can be explored. In this session we will focus
on the basic matplotlib figure. In particular, lineplots, barcharts and histograms will
be of interest.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


data_GER = pd.read_csv('UR_GER.csv')
data_JPN = pd.read_csv('UR_JPN.csv')
data_AUS = pd.read_csv('UR_AUS.csv')
data_KOR = pd.read_csv('UR_KOR.csv')
```

```python
# Unefficient way
data_GER.set_index('TIME', inplace=True)
data_JPN.set_index('TIME', inplace=True)
data_AUS.set_index('TIME', inplace=True)
data_KOR.set_index('TIME', inplace=True)
```

```python
# EXERICSE
UR=[data_GER, data_JPN, data_AUS, data_KOR]

for data in UR:
    data.set_index('TIME', inplace=True)
```

# Basic Plotting: One Axe

```python
fig, ax = plt.subplots()

# Define size of figure
fig.set_size_inches(10, 5)

# Customize data mapping appearance
ax.plot(data_GER,
        marker="o",
        color='g',
        linestyle='--',
        linewidth=5,
        label='UR GER')
ax.plot(data_JPN, '*b-', label='UR JPN')

# Customize axes labels
ax.set_xlabel('Years')
ax.set_ylabel('Percentage')

# Customize title
ax.set_title("OECD Unemployment Rates", fontsize=20)

# Add a grid
plt.grid()

# Add legend (use label= method in .plt() instead of passing a list to .legend()
plt.legend()

# Save plot
plt.savefig('plot.png')  # saved into main directory

plt.show()
```

# Basic Plotting: Multiple Axes

```python
fig, ax = plt.subplots(2, 2, sharey=True, sharex=True)         # creates array of axes, here: 2 x 2d arrays
ax[0, 0].plot(data_GER, linestyle = '--')
ax[1, 0].plot(data_JPN, color='g')
ax[0, 1].plot(data_AUS, linestyle='None', marker='x')
ax[1, 1].plot(data_KOR, linewidth=8)
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, sharey=True, sharex=True)         # creates array of axes, here: 1 x 2d arrays
ax[0].plot(data_GER, linestyle = '--')
ax[1].plot(data_AUS, linestyle='None', marker='x')
plt.show()
```

# Basic Plotting: Histogram

```python
hist_data = [14, 32, 23, 44, 5, 6, 7, 8, 9, 10, 33, 44, 55, 66, 35, 123]

my_bins = [10, 20, 30, 40, 50, 60, 100]
my_bins2 = [10, 20, 30, 40, 50, 60]


# ======  Customize bins

# by number of bins
plt.hist(hist_data, bins=5, edgecolor='black')
plt.show()

# by personal range
plt.hist(hist_data, bins=my_bins)
plt.show()

plt.hist(hist_data, bins=my_bins2)  # Excluding data by range
plt.show()
```

```python
# ======  Customize scale

# log scale
plt.hist(hist_data, bins=my_bins, log=True)
plt.show()
```

```python
# ======  Customize with markers

mean = np.mean(hist_data)

# Adding horizontal and vertical markers
plt.hist(hist_data, bins=my_bins)
plt.axvline(mean, color='red', label='Mean', linewidth=3)
plt.axhline(1.5, color='green', label='Mean', linewidth=3)
plt.legend()
plt.show()
```

```python
# Adding multiple vertical lines
plt.hist(hist_data, bins=my_bins)
plt.axvline(mean, color='red', label='Mean', linewidth=3)
plt.axvline(60, color='red', label='60', linewidth=3)
plt.legend()
plt.show()
```
